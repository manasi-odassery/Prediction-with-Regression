```python
import pandas as pd
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
del_df = pd.read_csv(r"C:/Manasi/Topmentor/Assignments/Assignment/Project - 1&2_ Simple_Reg/Project - 1&2_ Simple_Reg/delivery_time.csv")
sal_df = pd.read_csv(r"C:/Manasi/Topmentor/Assignments/Assignment/Project - 1&2_ Simple_Reg/Project - 1&2_ Simple_Reg/Salary_Data.csv")
```

```python
# Plot for Delivery Time Data
plt.figure(figsize=(10, 4))
sns.scatterplot(data=del_df, x='Sorting Time', y='Delivery Time')
plt.title('Delivery Time vs Sorting Time')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.show()
```

```python
# Plot for Salary Data
plt.figure(figsize=(10, 4))
sns.scatterplot(data=sal_df, x='YearsExperience', y='Salary')
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
X_del = del_df[['Sorting Time']]
y_del = del_df['Delivery Time']
```

```python
# Splitting the data into training and testing sets
X_train_del, X_test_del, y_train_del, y_test_del = train_test_split(X_del, y_del, test_size=0.2, random_state=42)
```

```python
# Creating the regression model for delivery time
model_del = LinearRegression()
model_del.fit(X_train_del, y_train_del)
```

```python
predictions_del = model_del.predict(X_test_del)
```

```python
# Evaluating the Delivery model
mse_del = mean_squared_error(y_test_del, predictions_del)
r2_del = r2_score(y_test_del, predictions_del)
```

```python
print(f'Delivery Time Prediction Model - MSE: {mse_del}, R2: {r2_del}')
```

```python
X_sal = sal_df[['YearsExperience']]
y_sal = sal_df['Salary'] 
```

```python
# Splitting the data into training and testing sets
X_train_sal, X_test_sal, y_train_sal, y_test_sal = train_test_split(X_sal, y_sal, test_size=0.2, random_state=42)
```

```python
# Creating the regression model for salary
model_sal = LinearRegression()
model_sal.fit(X_train_sal, y_train_sal)
```

```python
predictions_sal = model_sal.predict(X_test_sal) 
```

```python
# Evaluating the Salary model
mse_sal = mean_squared_error(y_test_sal, predictions_sal)
r2_sal = r2_score(y_test_sal, predictions_sal)
```

```python
print(f'Salary Prediction Model - MSE: {mse_sal}, R2: {r2_sal}')
```
